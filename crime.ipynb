The first metric is crime. In this case, we are trying to build a measurement of the security degree of the community.                  
The measurement can be divided into two parts: the probability of encounter crime and how severe the crime is.                      
For the first part, we need to figure out the frequency of each type of crime and proportion of criminal in the whole population of the community.                
For the second part, we need to have a rank of the severe degree of differernt types of crime, and figure a weight to quantify the potential harm of these crimes. The weight will be represented by numbers between 0 and 100, where 100 means death.                     
And the final measurement will be built like this:         
            
1.Find out the proportion of criminal in the poplulation of community, namely the probability of encountering a criminal in the community.( a%)      
2.Use the frequency of different types of crime to predict what would be the next action of this criminal. (m% for murder, n% for robber and so on ) we assume that the type of crime won't convert to each other and even the slightest crime is intolerable.           
3.Times the fequency by the weight and add them together.(100m% for murder+70n% for robber...) The result will be taken as a criminal's personal
dangerous degree.        
4.Times the criminal dangerous degree by the probability of encountering a criminal, the result will be the community dangerous degree.                   
                
The lower the community dangerous degree, the better the security degree of the community is.



```python
import pandas as pd
import numpy as np
import geopandas as gpd
crime = pd.read_csv("crime.csv")
list1=pd.DataFrame(crime.iloc[0:,[0,2,6,7]])
list1.head(20)
```

```python
a=list1.iloc[0:,2]
b=list1.iloc[0:,3]
c=float(list1.iloc[0,1].replace(',',''))

#this list shows the probability of encountering a criminal in the community.
# PEC=probability of encountering a criminal
list2={'neighborhood':crime.iloc[0:,0],'PEC':(a+b)/c}

x=pd.DataFrame(list2)
x1=x.sort_values(by="PEC").head()
x1
```

Above are five communities that you are least likely to encounter a criminal.

```python
a1=crime.iloc[0:,8]
a2=crime.iloc[0,9]
a3=crime.iloc[0,10]
a4=crime.iloc[0,11]
a5=crime.iloc[0,12]
a6=crime.iloc[0,13]
a7=crime.iloc[0,14]
k=a1+a2+a3+a4+a5+a6+a7
#PDD=personal dangerous degree
PDD=100*a1/k+95*a2/k+90*a3/k+90*a4/k+95*a5/k+90*a6/k+90*a7/k
list3={'neighborhood':crime.iloc[0:,0],'PDD':PDD}
y=pd.DataFrame(list3)
y1=y.sort_values(by="PDD").head()
y1
```

Above are five communities that the criminals inside are "least" dangerous.
